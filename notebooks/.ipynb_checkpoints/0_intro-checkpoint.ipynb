# Наш первый блокнот

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
math.sqrt(4)
```

```python
x_list = np.arange(-2.0, 2.1, 0.1)
x_list
```

```python
def sqr(x):
    return x**2
```

```python
y_list = [sqr(x) for x in x_list]
y_list
```

```python
plt.plot(x_list, y_list, 'b.')
plt.show() 
```

```python
plt.plot(x_list, y_list, 'g-')
plt.show() 
```
